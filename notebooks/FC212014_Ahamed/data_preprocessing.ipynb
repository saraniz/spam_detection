```python
import pandas as pd

df = pd.read_csv('../../data/maildata.csv')
df.head()
```

```python
df.info()
```

```python
def encode_categories(df, category_column='Category'):
    df_encoded = df.copy()
    df_encoded['Target'] = df_encoded[category_column].map({'spam': 1, 'ham': 0})
    return df_encoded

df= encode_categories(df)
df.head()
```

```python
df.drop(columns='Category')
df.head()
```

```python
import os
from sklearn.model_selection import train_test_split


os.makedirs('splits', exist_ok=True)


train_val_df, test_df = train_test_split(df, test_size=0.2, random_state=42)

train_val_df.to_csv('splits/train.csv', index=False)
test_df.to_csv('splits/test.csv', index=False)
```
