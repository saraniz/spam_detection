# Exploratory Data Analysis (EDA)

```
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from wordcloud import WordCloud
```

```
df = pd.read_csv('/mnt/data/maildata.csv')
df.head()
```

```
df['Category'].value_counts().plot(kind='bar', title='Class Distribution', xlabel='Category', ylabel='Count')
plt.show()
```

```
df['Message Length'] = df['Message'].apply(len)
df['Message Length'].hist(by=df['Category'], bins=50, figsize=(10,4))
```

```

spam_words = ' '.join(list(df[df['Category'] == 'spam']['Message']))
ham_words = ' '.join(list(df[df['Category'] == 'ham']['Message']))
spam_wc = WordCloud(width=600, height=400).generate(spam_words)
ham_wc = WordCloud(width=600, height=400).generate(ham_words)

plt.figure(figsize=(12,6))
plt.subplot(1,2,1)
plt.imshow(spam_wc)
plt.title('Spam Word Cloud')
plt.axis('off')

plt.subplot(1,2,2)
plt.imshow(ham_wc)
plt.title('Ham Word Cloud')
plt.axis('off')
plt.show()
```
