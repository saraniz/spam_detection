```python
import seaborn as sns
import matplotlib.pyplot as plt

df['Category'].value_counts().plot(kind='bar', color=['skyblue', 'salmon'])
plt.title("Count of Ham vs Spam")
plt.xlabel("Category")
plt.ylabel("Count")
plt.show()
```

```python
df['Message_Length'] = df['Message'].apply(len)

sns.histplot(data=df, x='Message_Length', hue='Category', bins=50, kde=True)
plt.title("Message Length Distribution")
plt.xlabel("Length")
plt.ylabel("Count")
plt.show()
```

```python
from wordcloud import WordCloud

spam_words = ' '.join(df[df['Category'] == 'spam']['Message'])
ham_words = ' '.join(df[df['Category'] == 'ham']['Message'])

# Spam word cloud
wc = WordCloud(width=600, height=400).generate(spam_words)
plt.imshow(wc, interpolation='bilinear')
plt.axis('off')
plt.title('Spam Word Cloud')
plt.show()

# Ham word cloud
wc = WordCloud(width=600, height=400).generate(ham_words)
plt.imshow(wc, interpolation='bilinear')
plt.axis('off')
plt.title('Ham Word Cloud')
plt.show()
```
